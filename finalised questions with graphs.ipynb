# Handling missing values

```python
# First, let's check for missing values in each column to assess which columns need cleaning
missing_values = df.isnull().sum().sort_values(ascending=False)

# Display columns with missing values and their counts
missing_values[missing_values > 0]
```

```python
#Drop columns which has signficant missing values
Col_drop =['Street Suffix','Street Prefix','End_Date_Time']
```

```python
df =df.drop(columns=Col_drop)
```

```python
df
```

```python
columns_to_update = ['Dispatch Date / Time', 'Start_Date_Time']

# Convert the columns to datetime64
for column in columns_to_update:
    df[column] = pd.to_datetime(df[column], errors='coerce')

# Replace 'Not Defined' with NaT
df[columns_to_update] = df[columns_to_update].replace('Not Defined', pd.NaT)

df['Start_Date_Time'] = pd.to_datetime(df['Start_Date_Time'], errors ='coerce')
df['Dispatch Date / Time'] = pd.to_datetime(df['Dispatch Date / Time'], errors ='coerce')
df['Dispatch Date / Time'] = pd.to_datetime(df['Dispatch Date / Time'],format ='%Y-%m-%d %H:%M:%S.%f')
df['Start_Date_Time'] = pd.to_datetime(df['Start_Date_Time'],format ='%Y-%m-%d %H:%M:%S.%f')
df['Start_Date_Time'] = df['Start_Date_Time'].fillna(pd.NaT)
df['Dispatch Date / Time'] = df['Dispatch Date / Time'].fillna(pd.NaT)
df.fillna(pd.NaT, inplace=True)
```

```python
# Again checking missing values
missing_data = df.isnull().sum()
print(missing_data)
```

```python
df.info()
```

# Converting date/time and adding formats

```python
df['Dispatch Date / Time'] = df['Dispatch Date / Time'].fillna(df['Start_Date_Time'])
df.dropna(subset=['Crime Name1', 'Crime Name2', 'Crime Name3'],inplace=True)
crime_columns = ['Police District Name','Block Address','City','Zip Code','Sector','Beat','PRA','Address Number','Street Name','Street Type']
df[crime_columns] = df[crime_columns].fillna('Not Defined')
df['Year'] = df['Dispatch Date / Time'].dt.year
df['DayOfWeek'] = pd.to_datetime(df['Dispatch Date / Time']).dt.day_name()
df['Hour'] = pd.to_datetime(df['Dispatch Date / Time']).dt.hour
df['Month'] = pd.to_datetime(df['Dispatch Date / Time']).dt.month
#group by function for season
```

```python
# Again checking missing values
missing_data = df.isnull().sum()
print(missing_data)
```

# Q1: What are the incidence rates for each category of crime in Montgomery County annually?

```python
# Define crime count as we will need it to show number of crimes
crime_counts = df.groupby(['Year', 'Crime Name1']).size().reset_index(name='Counts')

# Plot the graph with specified data
plt.figure(figsize=(14, 8))
sns.barplot(data=crime_counts, x='Year', y='Counts', hue='Crime Name1')
plt.title('Incidence Rates for Each Category of Crime in Montgomery County Annually')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.legend(title='Crime Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Group the dataset by Year and Month for analysis
monthly_incidents = df.groupby(['Year', 'Month']).size().unstack(fill_value=0)

years = monthly_incidents.index
fig, axes = plt.subplots(nrows=len(years), ncols=1, figsize=(10, 12), sharex=True)

# Create a subplot graph and customise for readability
for i, year in enumerate(years):
    axes[i].plot(monthly_incidents.columns, monthly_incidents.loc[year], marker='o', color='black', linestyle='-')
    axes[i].scatter(monthly_incidents.columns, monthly_incidents.loc[year], color='red')
    axes[i].set_title(f'Total Incidents per Month in {year}', fontsize=14)
    axes[i].grid(True)
    
    min_val = monthly_incidents.loc[year].min()
    max_val = monthly_incidents.loc[year].max()
    y_ticks = [min_val, (min_val + max_val) // 2, max_val] 
    axes[i].set_yticks(y_ticks)

plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
                                       'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=0)
plt.xlabel('Month', fontsize=12)
plt.tight_layout()
plt.show()
```

# Q2. What are the most common offenses based on offense codes, and how do these top offenses contribute to the overall crime landscape? (Top 10 Crimes by Offense Code for crime name2)

# Q4. What are the seasonal patterns in crime rates for different types of crimes in Montgomery County?  

```python
# Filter out "All Other Offenses" and filter data for the top 10 crime types by 'Crime Name2' and group by 'Month' and 'Crime Name2'
top_10_crime_types = df['Crime Name2'].value_counts().head(11).index.difference(['All Other Offenses'])
filtered_df = df[df['Crime Name2'].isin(top_10_crime_types)]
monthly_counts = filtered_df.groupby(['Month', 'Crime Name2']).size().unstack(fill_value=0)

# Plot updated graph
plt.figure(figsize=(16, 10))
sns.heatmap(data=monthly_counts, annot=True, fmt='d', cmap='coolwarm', linewidths=0.3, linecolor='white',
            cbar_kws={'label': 'Incident Count'}) 
plt.title('Monthly Incident Counts for Top Crime Types in Montgomery County (Excluding "All Other Offenses")', fontsize=18)
plt.xlabel('Crime Type', fontsize=14)
plt.ylabel('Month', fontsize=14)
plt.yticks(ticks=np.arange(12) + 0.5, labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=0, fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Prepare data for the box plot
plt.figure(figsize=(16, 10))
sns.boxplot(data=filtered_df, x='Crime Name2', y='Month', orient='v')

# Calculate and annotate the mean for each crime type
means = filtered_df.groupby('Crime Name2')['Month'].mean()
for i, crime in enumerate(filtered_df['Crime Name2'].unique()):
    mean_value = means[crime]
    plt.text(i, mean_value, f'{mean_value:.1f}', ha='center', va='bottom', color='black', fontsize=10)

# Customize the plot
plt.title('Distribution of Monthly Incidents for Top Crime Types in Montgomery County (Excluding "All Other Offenses")', fontsize=18)
plt.xlabel('Crime Type', fontsize=14)
plt.ylabel('Month', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.tight_layout()
plt.show()
```

# Q6 What times of day have the highest frequency of crimes in Montgomery County? 

```python
# Create a 'Hour' column if not already done by extracting hour from 'Dispatch Date / Time' datetime column
df['Hour'] = df['Dispatch Date / Time'].dt.hour

# Plot histogram using histplot to show the frequency of incidents by hour
plt.figure(figsize=(12, 6))
sns.histplot(df['Hour'], bins=24, kde=True, color='coral')
plt.xlabel('Hour of the Day', fontsize=14)
plt.ylabel('Number of Incidents', fontsize=14)
plt.title('Crime Frequency by Hour of the Day in Montgomery County', fontsize=16)
plt.xticks(range(0, 24), [f'{i}:00' for i in range(0, 24)], rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.tight_layout()

# Identify and annotate the peak hour
crime_by_hour = df['Hour'].value_counts().sort_index()
max_incidents = crime_by_hour.max()
peak_hour = crime_by_hour.idxmax()
plt.annotate(f'Peak Hour: {peak_hour}:00\n({max_incidents} incidents)', 
             xy=(peak_hour, max_incidents), xytext=(peak_hour+1, max_incidents+10),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontsize=8, bbox=dict(facecolor='white', alpha=0.6))

plt.show()
```

```python
# Filter the data for crimes occurring at 15:00 and group by 'Crime Name3'
peak_hour_data = df[df['Hour'] == 15]
top_crimes_at_peak = peak_hour_data['Crime Name3'].value_counts().head(10).reset_index() 
top_crimes_at_peak.columns = ['Crime Type', 'Count'] 

# Sort the data and plot a line plot with the data 
top_crimes_at_peak_sorted = top_crimes_at_peak.sort_values(by='Count', ascending=False)

# Improving readability by customising line and markers and annotating the highest and lowest points of crime
plt.figure(figsize=(14, 8))
plt.plot(top_crimes_at_peak_sorted['Crime Type'], top_crimes_at_peak_sorted['Count'], 
         linestyle='-', color='black', linewidth=2.5, label='Crime Count') 
plt.scatter(top_crimes_at_peak_sorted['Crime Type'], top_crimes_at_peak_sorted['Count'], 
            color='red', s=100, label='Data Points')  

max_index = top_crimes_at_peak_sorted['Count'].idxmax()
min_index = top_crimes_at_peak_sorted['Count'].idxmin()
plt.annotate(f'Peak: {top_crimes_at_peak_sorted["Count"].loc[max_index]} incidents', 
             xy=(top_crimes_at_peak_sorted['Crime Type'].loc[max_index], top_crimes_at_peak_sorted['Count'].loc[max_index]), 
             xytext=(max_index, top_crimes_at_peak_sorted['Count'].loc[max_index] + 200),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontsize=10, bbox=dict(facecolor='white', alpha=0.6))

plt.annotate(f'Low: {top_crimes_at_peak_sorted["Count"].loc[min_index]} incidents', 
             xy=(top_crimes_at_peak_sorted['Crime Type'].loc[min_index], top_crimes_at_peak_sorted['Count'].loc[min_index]), 
             xytext=(min_index, top_crimes_at_peak_sorted['Count'].loc[min_index] - 200),
             arrowprops=dict(facecolor='red', arrowstyle='->'),
             fontsize=10, bbox=dict(facecolor='white', alpha=0.6))


plt.title('Top 10 Crimes at Peak Hour (15:00)' )
plt.xlabel('Crime Type')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45, ha='right')  
plt.grid(True, linestyle='--', linewidth=0.5)  
plt.tight_layout()
plt.legend()
plt.show()
```

## Q8. How does the number of victims vary across different Street Name for crimes involving more than one victim, and which Street Name are more likely to have higher number of victims? 

```python
# Create 'Place Group' column based on 'Place Category', to separate commercial, residential and other
df['Place Group'] = 'Other' 

# Define conditions for 'Residential' and 'Commercial'
residential_conditions = df['Place Category'].str.contains('Residential', case=False, na=False)
commercial_conditions = df['Place Category'].str.contains('Commercial|Business|Shop|Store', case=False, na=False)

# Apply conditions to assign appropriate labels
df.loc[residential_conditions, 'Place Group'] = 'Residential'
df.loc[commercial_conditions, 'Place Group'] = 'Commercial'

# Check if 'Place Group' column has been created successfully
print(df['Place Group'].value_counts())
```

# Q9. How has the number of victims vary over time for the top 10 most active beats, and are there any noticeable trends or patterns in the data?

```python
# Add a grid for better readability
sns.set_theme(style="whitegrid")

# Filter data for the top 10 beats with the highest total victim count
top_beats = df.groupby('Beat')['Victims'].sum().nlargest(10).index
df_filtered = df[df['Beat'].isin(top_beats)]

# Plot a stacked bar chart showing the number of victims per year with filtered data and customise further for readability
plt.figure(figsize=(14, 8))
df_grouped = df_filtered.groupby(['Year', 'Beat'])['Victims'].sum().unstack(fill_value=0)
df_grouped.plot(kind='bar', stacked=True, colormap='tab10', ax=plt.gca())

for i, year in enumerate(df_grouped.index):
    total_victims = df_grouped.loc[year].sum()
    plt.text(i, total_victims + 500, f'{total_victims}', ha='center', va='bottom')

plt.title('Number of Victims Over Time for Top 10 Most Active Beats', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Victims', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='Beat', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Plot a lie cart for Top 5 Beats by year
plt.figure(figsize=(14, 8))
top_5_beats = df_filtered['Beat'].value_counts().head(5).index

for beat in top_5_beats:
    subset = df_filtered[df_filtered['Beat'] == beat].groupby('Year')['Victims'].sum()
    plt.plot(subset.index, subset.values, marker='o', label=f'Beat {beat}', linewidth=2)

plt.xticks(subset.index, subset.index.astype(int), rotation=45)

plt.title('Crime Trends by Year for Top 5 Beats', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Victims', fontsize=14)
plt.grid(True)
plt.legend(title='Beat', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

# Q10. How does the distribution of victims vary across different months and which types of crimes (crime name two) are most prevalent in each month?

```python
# Group by 'Crime Name2' to get the total victim count for each crime type
top_crimes = df.groupby('Crime Name2')['Victims'].sum().nlargest(10).index


# Filter the dataset to include only the top 10 crimes
df_top_crimes = df[df['Crime Name2'].isin(top_crimes)]

# Group by 'Month' and 'Crime Name2' for the filtered dataset, then get the victim count
monthly_victim_distribution = df_top_crimes.groupby(['Month', 'Crime Name2'])[('Victims')].sum().unstack().fillna(0)

# Plotting
plt.figure(figsize=(14, 8))
ax = monthly_victim_distribution.plot(kind='bar', stacked=True, colormap='tab20', ax=plt.gca())
plt.title("Monthly Distribution of Victims by Top 10 Crime Types")
plt.xlabel("Month")
plt.ylabel("Number of Victims")
plt.xticks(rotation=45)
plt.legend(title="Crime Type", bbox_to_anchor=(1.05, 1), loc='upper left')

# Add text annotations to show the number of victims in each stacked bar segment
for container in ax.containers:
    # Add labels to each bar segment
    ax.bar_label(container, label_type='center', fontsize=8, color='black')

plt.show()
```
